```python
# importing all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Reading dataset
path="http://bit.ly/w-data"
data=pd.read_csv(path)
```

```python
#reading first 5 values
data.head()
```

```python
#reading last 5 values
data.tail()
```

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours Vs Percentage')
plt.xlabel('Studied hours')
plt.ylabel('Percentage score')
plt.show()
```

```python
# Dividing data into inputs and outputs
X=data.iloc[:,:-1].values
Y=data.iloc[:,1].values
```

```python
X
```

```python
Y
```

```python
# splitting the data into training sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train, Y_test=train_test_split(X.reshape(-1,1),Y,test_size=0.2)
#Scikit doesn't work with scalars(single value). Hence, reshaping its dimension.
#It expects a shape (m x n) 
'''where m-> Number of features 
         n-> Number of Observations'''
```

```python
# training the algorithm
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)

print("Training completed successfully!")
```

```python
# Plotting the linear regression line
line=regressor.coef_*X + regressor.intercept_
plt.scatter(X,Y)
plt.plot(X,line);
plt.show()
```

```python
# Predictions
print(X_test) #testing data(in hours)
Y_pred=regressor.predict(X_test) #predicting scores
```

```python
#comparing actual and predicted values
df=pd.DataFrame({'Actual':Y_test,'Predicted':Y_pred})
print(df)
```

```python
#testing with our data
hours=9.25 #given testcase
own_pred=regressor.predict([[hours]])
```

```python
print("Hours: {}".format(hours))
print("Predicted Score: {}".format(own_pred[0]))
```

```python
#Evaluating our algorithm
from sklearn import metrics
```

```python
# mean absolute error
print('Mean Absolute Error:',metrics.mean_absolute_error(Y_test,Y_pred))
```

```python
# mean square error
print('Mean Square Error:',metrics.mean_squared_error(Y_test,Y_pred))
```
