```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading data
iris_df=pd.read_csv('Iris.csv')
```

```python
# last five values from dataset
iris_df.tail()
```

```python
# columns in the dataset
iris_df.columns
```

```python
# dividing the data into two subsets
X=iris_df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
Y=iris_df['Species'].values
```

```python
# To find: optimal number of clusters
# Elbow Method
# importing KMeans
from sklearn.cluster import KMeans
sswc=[] #creating an empty list
```

```python
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X)
    sswc.append(kmeans.inertia_) 
```

```python
#plotting the graph
plt.plot(range(1,10),sswc)
plt.title('ELBOW METHOD')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
# Kmeans clustering Algorithm
kmeans_3=KMeans(n_clusters=3)
kmeans_3.fit(X)
print(kmeans_3.cluster_centers_)
pred_clusters=kmeans_3.fit_predict(X)
print(pred_clusters)

```

```python
# Clustering Visualization
plt.scatter(X[:,0],X[:,1],c=pred_clusters,cmap='viridis')
plt.show()
```
